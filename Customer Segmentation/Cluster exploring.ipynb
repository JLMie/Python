### EXPLORING CLUSTERS

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
autodf_clusters = pd.read_csv('autodf_clusters.csv')
```

```python
# Spending_Score Spending score of the customer.
# Family_Size Number of family members for the customer (including the customer).
# Var_1 Anonymised Category for the customer.
# Segmentation (target) Customer Segment of the customer.
```

```python
autodf_clusters.head(5)
```

```python
autodf_clusters.shape
```

## CLUSTER 0

```python
cluster_0 = autodf_clusters[autodf_clusters['clusters'] == 0]
cluster_0 = pd.DataFrame(cluster_0)
cluster_0.describe().T
```

```python
def exploring_cluster(cluster_0, column_name):
    result = pd.DataFrame({
        'Percentage': cluster_0[column_name].value_counts(normalize=True).mul(100).round(2),
        'Count': cluster_0[column_name].value_counts()
    })
    
    return result
```

```python
def explore_all_columns(cluster_0):
    for column_name in cluster_0.columns:
        result = exploring_cluster(cluster_0, column_name)
        print(f">>>>> {column_name.upper()}")
        print(result)

explore_all_columns(cluster_0)
```

```python
def plot_numeric(column_name):
    fig, ax = plt.subplots(1, 2)
#    fig.suptitle(f"{column_name} Distribution", fontsize=15)

    sns.histplot(cluster_0[column_name], ax=ax[0])
    ax[0].set_title("Distribution")
    
plot_numeric("Age")
plot_numeric("Work_Experience")
plot_numeric("Work_Experience_to_Age_Ratio")
plot_numeric("weekly_average_km_values")
plot_numeric("numbers_of_occupant")
```
