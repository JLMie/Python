```python
import numpy as np
import pandas as pd

# Define the total number of rows
total_rows = 10600

# Calculate the number of values for each range in "weekly_average_km_values"
range_10_20 = int(total_rows * 0.05)
range_21_40 = int(total_rows * 0.1)
range_41_60 = int(total_rows * 0.21)
range_61_90 = int(total_rows * 0.32)
range_91_120 = int(total_rows * 0.19)
range_121_140 = int(total_rows * 0.07)
range_141_180 = int(total_rows * 0.06)

# Create the range arrays for "weekly_average_km_values"
range_values_10_20 = np.random.randint(10, 21, size=range_10_20)
range_values_21_40 = np.random.randint(21, 41, size=range_21_40)
range_values_41_60 = np.random.randint(41, 61, size=range_41_60)
range_values_61_90 = np.random.randint(61, 91, size=range_61_90)
range_values_91_120 = np.random.randint(91, 121, size=range_91_120)
range_values_121_140 = np.random.randint(121, 141, size=range_121_140)
range_values_141_180 = np.random.randint(141, 181, size=range_141_180)

# Combine all the range arrays
weekly_average_km_values = np.concatenate([
    range_values_10_20,
    range_values_21_40,
    range_values_41_60,
    range_values_61_90,
    range_values_91_120,
    range_values_121_140,
    range_values_141_180
])

# Shuffle the values
np.random.shuffle(weekly_average_km_values)

# Create the range arrays for "type_of_road_values"
type_of_road_values = np.random.choice(
    ["forest", "village", "town", "city"],
    size=total_rows,
    p=[0.13, 0.19, 0.3, 0.38]
)

# Create the range arrays for "numbers of occupant"
numbers_of_occupant = np.random.choice(
    [1, 2, 3, 4, 5],
    size=total_rows,
    p=[0.36, 0.3, 0.18, 0.1, 0.06]
)

# Create the DataFrame
df = pd.DataFrame({
    "weekly_average_km": weekly_average_km_values,
    "type_of_road": type_of_road_values,
    "numbers_of_occupant": numbers_of_occupant
})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_numeric(column_name):
    fig, ax = plt.subplots(1, 2)
    fig.suptitle(f"{column_name} Distribution", fontsize=15)

    sns.histplot(df[column_name], ax=ax[0])
    ax[0].set_title("Distribution")

    sns.boxplot(df[column_name], ax=ax[1])
    ax[1].set_title("Box Plot")

    plt.tight_layout()
    plt.show()

plot_numeric("weekly_average_km_values")
plot_numeric("numbers_of_occupant")
```

```python
import matplotlib.pyplot as plt

def plot_frequency(column_name):
    plt.figure(figsize=(4, 3))
    df[column_name].value_counts().plot.bar(color="purple")
    plt.title(f"{column_name} Frequency", fontsize=15)
    plt.show()

plot_frequency("type_of_road")
```
